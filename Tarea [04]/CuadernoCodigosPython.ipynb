### 1. Use el método de bisección para encontrar soluciones precisas dentro de $10^-2$ para $x^3-7x^2+14x-6=0$ en cada intervalo 

#### a.[0,1]

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para la cual se aplicará el método de bisección
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

# Método de Bisección
def bisection_method(func, a, b, tol=1e-2, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None
    
    iteration = 0
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        if func(midpoint) == 0:
            return midpoint  # Solución exacta encontrada
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    
    return (a + b) / 2  # Retorna la solución aproximada

# Definición del intervalo
a, b = 0, 1

# Ejecutar el método de bisección
root = bisection_method(f, a, b)

# Generar la gráfica de la función en el intervalo [0, 1]
x_values = np.linspace(a, b, 500)
y_values = f(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = x^3 - 7x^2 + 14x - 6", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle="--", label=f"Root ≈ {root:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función y aproximación de la raíz por el método de bisección")
plt.legend()
plt.grid(True)
plt.show()
```

#### b. Use el metodo de bisección para encontrar una aproximación dentro de $10^-3$ para un valor en [-2, 0] con $x^2-1=e^{1-x^2}$

```python
import numpy as np

# Definimos la función f(x) = x^2 - 1 - exp(1 - x^2)
def f_bisection(x):
    return x**2 - 1 - np.exp(1 - x**2)

# Método de Bisección modificado para una precisión de 10^-3
def bisection_method_precise(func, a, b, tol=1e-3, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None
    
    iteration = 0
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        if func(midpoint) == 0:
            return midpoint  # Solución exacta encontrada
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    
    return (a + b) / 2  # Retorna la solución aproximada

# Definición del intervalo
a, b = -2, 0

# Ejecutar el método de bisección con la nueva función y precisión
root_precise = bisection_method_precise(f_bisection, a, b)
print("Aproximación de la raíz:", root_precise)
```

#### 1. Un abrevadero de longitud 𝐿 tiene una sección transversal en forma de semicírculo con radio 𝑟. (Consulte la figura adjunta.) Cuando se llena con agua hasta una distancia ℎ a partir de la parte superior, el volumen 𝑉 de agua es:
![](imagen3.png)
#### Suponga que 𝐿 = 10 𝑐𝑚, 𝑟 = 1 𝑐𝑚 y 𝑉 = 12.4 $𝑐𝑚^{3}$. Encuentre la profundidad del agua en el abrevadero dentro de 0.01 𝑐𝑚.

```python
import math
from scipy.optimize import fsolve

# Datos dados
L = 10      # Longitud en cm
r = 1       # Radio en cm
V = 12.4    # Volumen en cm^3

# Definimos la función para el volumen en función de h
def volumen_func(h):
    # Calcula el volumen basado en la fórmula dada
    term1 = 0.5 * math.pi * r**2
    term2 = r**2 * math.asin(h / r)
    term3 = h * math.sqrt(r**2 - h**2)
    volumen_calculado = L * (term1 - term2 - term3)
    
    # La función devuelve la diferencia entre el volumen deseado y el calculado
    return volumen_calculado - V

# Estimación inicial para h (por ejemplo, la mitad del radio)
h_inicial = r / 2

# Usamos fsolve para encontrar el valor de h
h_solucion, = fsolve(volumen_func, h_inicial)

# Mostramos el resultado
print(f"La profundidad del agua en el abrevadero es aproximadamente {h_solucion:.4f} cm")
```

#### 2. Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa 𝑚 cae desde una altura $s_0$ y que la altura del objeto después de 𝑡 segundos es:
![](imagen4.png)
#### donde g = 9.81 m/s y k representa el coeficiente de la resistencia del aire en NS/m. Suponga $S_0=300m$. m=0.25kg y k=0.1 NS/m. Encuentre, dentro de 0.01 segundos, el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import numpy as np
from scipy.optimize import fsolve

# Parámetros dados
g = 9.81  # aceleración de la gravedad en m/s^2
m = 0.25  # masa en kg
k = 0.1   # coeficiente de resistencia en Ns/m
s0 = 300  # altura inicial en metros

# Definimos la función s(t) para la altura en función del tiempo
def s(t):
    return s0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - np.exp(-k * t / m))

# Encontramos el tiempo en que s(t) es aproximadamente 0
tiempo_inicial_estimado = 0.01  # tiempo inicial de búsqueda en segundos
tiempo_suelo = fsolve(s, tiempo_inicial_estimado)[0]

# Resultado
print("El tiempo en que el objeto golpea el piso es aproximadamente:", tiempo_suelo, "segundos")
```

##### 1. Use el teorema 2.1 para encontrar una cota para el número de iteraciones necesarias para lograr una aproximación con presición de $10^-4$ para la solución de $x^3-x-1=0$ que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximación para la raíz con este grado de precisión.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del teorema
a, b = 1, 2
epsilon = 1e-4

# Cálculo del número de iteraciones necesarias
N = np.ceil((np.log(b - a) - np.log(epsilon)) / np.log(2))
print("Número mínimo de iteraciones necesarias:", int(N))

# Definimos la función para la ecuación x^3 - x - 1 = 0
def f(x):
    return x**3 - x - 1

# Método de Bisección modificado para una precisión de 10^-4
def bisection_method_precise(func, a, b, tol=epsilon, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None, []
    
    iteration = 0
    midpoints = []  # Lista para almacenar los puntos medios
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        midpoints.append(midpoint)  # Guardar el punto medio
        if func(midpoint) == 0:
            return midpoint, midpoints  # Solución exacta encontrada
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    
    return (a + b) / 2, midpoints  # Retorna la solución aproximada

# Ejecutar el método de bisección con precisión de 10^-4
root_precise, midpoints = bisection_method_precise(f, a, b)
print("Aproximación de la raíz:", root_precise)

# Gráfica de la función y los puntos de bisección
x = np.linspace(0, 2, 100)  # Rango de x para la gráfica
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x³ - x - 1', color='blue')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Línea horizontal en y=0
plt.axvline(root_precise, color='red', linestyle='--', label='Raíz aproximada')
plt.scatter(midpoints, f(np.array(midpoints)), color='orange', label='Puntos medios', zorder=5)

# Etiquetas y título
plt.title('Método de Bisección')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-3, 3)  # Ajustar el rango del eje y para mejor visualización
plt.show()
```
